```python
#import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
#from sympy import symbols, diff
import sympy as sp
from sympy import simplify
```

Definimos las variables r_GC,  R_GC y z

```python
G, M_n, M_b, M_d, M_h, r_GC, r_n, r_b, r_s, R_GC, z_GC, a_d, b_d  = sp.symbols('G M_n M_b M_d M_h r_GC r_n r_b r_s R_GC z_GC a_d b_d', real=True)
```

```python
### Funciones ####
#Empezamos con el potencial del n√∫cleo
pot_nuc = -G*M_n/(r_GC + r_n)
sp.diff(pot_nuc, r_GC)
```

```python
# Calculamos la derivada de este gradiente de potencial para hallar el error asociado con el tiempo adaptativo (h)
der_segun_nuc = sp.diff(pot_nuc, r_GC)
simplify(sp.diff(der_segun_nuc,r_GC))
```

```python
#Potencial del bulbo
pot_bul = -G*M_b/(r_GC + r_b)
sp.diff(pot_bul, r_GC)
```

```python
# Calculamos la derivada de este gradiente de potencial para hallar el error asociado con el tiempo adaptativo (h)
der_segun_bul = sp.diff(pot_bul, r_GC)
simplify(sp.diff(der_segun_bul,r_GC))
```

```python
#Potencial del disco requiere derivada cruzada
pot_dis = -G*M_d/sp.sqrt(R_GC**2 + (a_d + sp.sqrt(z_GC**2 + b_d**2))**2)  
sp.diff(pot_dis, R_GC)
```

```python
sp.diff(pot_dis, z_GC)
```

```python
# Calculamos la derivada de este gradiente de potencial para hallar el error asociado con el tiempo adaptativo (h)

der_seg_dis_R = sp.diff(pot_dis, R_GC)
simplify(sp.diff(der_seg_dis_R,R_GC))
```

```python
# Calculamos la derivada de este gradiente de potencial para hallar el error asociado con el tiempo adaptativo (h)
der_seg_dis_z = sp.diff(pot_dis, z_GC)
simplify(sp.diff(der_seg_dis_z,z_GC))
```

```python
#Potencial del halo

pot_hal = -G*M_h*sp.log(1+(r_GC/r_s))/r_GC
#exp_pot_hal = (1+(r_GC/r_s))**(-G*M_h/r_GC)
sp.diff(pot_hal, r_GC)
#sp.diff(exp_pot_hal,r_GC)
```

```python
grad_pot_hal = -G*M_h*((r_s + r_GC)*sp.log(1+r_GC/r_s)-r_GC)/((r_s+r_GC)*r_GC**2)
```

## Ahora hallemos los diferenciales en cartesianas

```python
x,y,z  = sp.symbols('x y z', real=True)
```

###Primero para el n√∫cleo que coincide con el bulbo

```python
pot_nuc_cart = -G*M_n/(sp.sqrt(x**2+y**2+z**2) + r_n)
sp.diff(pot_nuc_cart, x)
```

```python
pot_dis_cart = -G*M_d/sp.sqrt((x**2+y**2) + (a_d + sp.sqrt(z**2 + b_d**2))**2)
sp.diff(pot_dis_cart,x) 
```

```python
pot_dis_cart = -G*M_d/sp.sqrt((x**2+y**2) + (a_d + sp.sqrt(z**2 + b_d**2))**2)
sp.diff(pot_dis_cart,z) 
```

```python
#Potencial del halo

pot_hal_cart = -G*M_h*sp.ln(1+(sp.sqrt(x**2+y**2+z**2)/r_s))*(x**2+y**2+z**2)**-1
sp.diff(pot_hal_cart,x)
```

```python
#Pedazo halo

Deri_halo_Car = -G*M_h*sp.log(1+(sp.sqrt(x**2+y**2+z**2)/r_s))
sp.diff(Deri_halo_Car,x)
```

```python
#segundo pedazo de halo
Deri2_halo_Car = -G*M_h/(x**2+y**2+z**2)
sp.diff(Deri2_halo_Car,x)
```

```python
#Radio para una masa total dada halo
#Primero resolvermos la integral simbolicamente
masa_halo = M_h*r_s/(r_GC + r_s)**2
sp.integrate(M_h*r_s/(r_GC + r_s)**2,r_GC)
```
